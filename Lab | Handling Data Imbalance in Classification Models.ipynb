# Before balancing the data

- Concatenate numerical and categorical back together again for your X dataframe. Designate the Target as y.

- Split the data into a training set and a test set.
- Split further into train_num and train_cat. Also test_num and test_cat.
- Scale the features either by using normalizer or a standard scaler. (train_num, test_num)
- Encode the categorical features using One-Hot Encoding or Ordinal Encoding. (train_cat, test_cat)
- fit only on train data transform both train and test
- again re-concatenate train_num and train_cat as X_train as well as test_num and test_cat as X_test
- Fit a logistic regression model on the training data.
- Check the accuracy on the test data.

```python
# Using only TARGET_B

data = donors_df.drop(columns=['TARGET_D'])
```

```python
data['TARGET_B'].value_counts()
```

```python
y = data['TARGET_B']
X = data.drop(['TARGET_B'], axis =1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
display(X_train.shape)
display(X_test.shape)
display(y_train.shape)
display(y_test.shape)
```

```python
train_num = X_train.select_dtypes(np.number)
test_num = X_test.select_dtypes(np.number)
train_num.head()
```

```python
y_train.head()
```

```python
# Scaling data

from sklearn.preprocessing import StandardScaler



transformer = StandardScaler().fit(train_num)
numerical_train = transformer.transform(train_num)
numerical_test = transformer.transform(test_num)
```

```python
train_num_scaled= pd.DataFrame(numerical_train, columns=train_num.columns)
test_num_scaled= pd.DataFrame(numerical_test, columns=test_num.columns)
train_num_scaled.head()
```

```python
# Encode the categorical features 

train_cat= X_train.select_dtypes(object)
test_cat= X_test.select_dtypes(object)
train_cat.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore', drop='first').fit(train_cat)
encoded_train_cat = encoder.transform(train_cat).toarray()
cols = encoder.get_feature_names_out(input_features=train_cat.columns)
train_cat_encoded = pd.DataFrame(encoded_train_cat, columns=cols)
train_cat_encoded.head()



encoded_test_cat = encoder.transform(test_cat).toarray()
cols = encoder.get_feature_names_out(input_features=test_cat.columns)
test_cat_encoded = pd.DataFrame(encoded_test_cat, columns=cols)
test_cat_encoded.head()
```

```python
#re-concatenate train_num and train_cat as 'X_train' as well as test_num and test_cat as 'X_test'

X_train_treated= pd.concat((train_num_scaled, train_cat_encoded), axis=1)
X_test_treated= pd.concat((test_num_scaled, test_cat_encoded), axis=1)
```

```python
#Fit a logistic regression model on the training data,and check the accuracy on the test data

from sklearn.linear_model import LogisticRegression
classification = LogisticRegression(random_state=0, solver='lbfgs',
                  multi_class='ovr')
classification.fit(X_train_treated, y_train)
```

```python
predictions = classification.predict(X_test_treated)
predictions
classification.score(X_test_treated, y_test)
```

```python
# The score is too high for the test set , or we can say should not be expected 
```

# Managing imbalance in the dataset

# Upsampling

```python
X_train_treated.head()
```

```python
y_train.head()
```

```python
# To resample we have to combine our y and x features together

train = pd.concat([X_train_treated, y_train.reset_index(drop=True)],axis=1)
train.head()
```

```python
train.shape
```

```python
train['TARGET_B'].value_counts()
```

```python
from sklearn.utils import resample

category_0 = train[train['TARGET_B'] == 0]
category_1 = train[train['TARGET_B'] == 1]
```

```python
category_1_upsampled = resample(category_1, 
                                  replace=True, 
                                  n_samples = len(category_0))
```

```python
print(category_0.shape)
print(category_1_upsampled.shape)
```

```python
train_upsampled = pd.concat([category_0, category_1_upsampled], axis=0)
```

```python
train_upsampled['TARGET_B'].value_counts()
```

```python
# spilt them again to x features and y target

y_train_oversamp = train_upsampled['TARGET_B']
X_train_oversamp = train_upsampled.drop('TARGET_B',axis = 1)
```

```python
#fit a logistic regression model 

classification_over = LogisticRegression(random_state=0, solver='lbfgs',
                  multi_class='ovr')
classification_over.fit(X_train_oversamp, y_train_oversamp)
```

```python
predictions2 = classification_over.predict(X_test_treated)
predictions2
classification_over.score(X_test_treated, y_test)
```

```python
# The score now is less than before the upsampling, and we can say more reasonable 
```
